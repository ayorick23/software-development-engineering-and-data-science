## Tarea 1:

Dada f(x)=3x^2-x+1 usa SymPy para evaluar su valor en puntos específicos. Usando la función subs() de SymPy, determina el valor de la función en x=2 y x=0.5.

```python
#Importación de librería SymPy
import sympy as sp
```

```python
#Declaración de variable simbólica x
x = sp.Symbol('x')
```

```python
#Declaración de función f(x)
fX = 3*x**2 - x + 1
```

```python
#Evaluación en la función
x1 = fX.subs(x, 2)
x2 = fX.subs(x, 0.5)
```

```python
#Impresión de resultados
print(f"f(2) = {x1}")
print(f"f(0.5) = {x2}")
```

## Tarea 2:

Represente individualmente la gráfica de las siguientes funciones polinómicas en el intervalo de -10 ≤x≤10 utilizando el comando plt.plot() de matplotlib.

a) f(x)=2x-5   
b) g(x)=2(x-1)^2+50  
c) h(x)=x^3-2x^2+x-1

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 400)
```

```python
fX = 2*x - 5
gX = 2*(x-1)**2 + 50
hX = x**3 - 3*x**2 + x - 1
```

```python
plt.figure(figsize=(8, 5))
plt.plot(x, fX, label = "f(x) = 2x - 5")
plt.plot(x, gX, label = "g(x) = 2(x-1)^2 + x - 1")
plt.plot(x, hX, label = ("h(x) = x^3 - 3x^2 + x - 1"))
plt.title("Graficas de funciones polinomicas")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid()
plt.show()
```

## Tarea 3:

Represente la gráfica de las siguientes funciones utilizando el comando plot() de SymPy

a) f(x)=(2x-1)/(3x-1)

b) g(x)=√(4x)^2-3x), utilice comando sqrt() sympy para la raíz cuadrada y luego ajuste el dominio de la función de -2 a 2 y rango de -0.5 a 3, utilizando los parámetros xlim y ylim en la función plot.

c) h(x)=|3-x|, utilice el comando abs() para el valor absoluto y luego ajuste el dominio de la función de 0 a 6 y rango de 0 a 3, utilizando los parámetros xlim y ylim en la función plot.

```python
import sympy as sp
import matplotlib.pyplot as plt
```

```python
x = sp.Symbol('x')
```

```python
fX = (2*x - 1)/(3*x - 1)
gX = sp.sqrt(4*x**2 - 3*x)
hX = sp.Abs(3 - x)
```

```python
sp.plot(fX, (x, -10, 10), title="Grafica de f(x)", xlabel="x", ylabel="f(x)")
sp.plot(gX, (x, -2, 2), title="Grafica de g(x)", xlabel="x", ylabel="g(x)", xlim=(-2, 2), ylim=(-0.5, 3))
sp.plot(hX, (x, 0, 6), title="Grafica de h(x)", xlabel="x", ylabel="h(x)", xlim=(0, 6), ylim=(0, 3))
```
