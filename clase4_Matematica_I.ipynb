```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 400)
```

```python
len(x)
```

```python
f = x**2 - 4*x + 3
g = -x**3 + 2*x**2 + x - 5
h = x**4 - 3*x**2 + 2

plt.plot(x, f, label ='f(x) = x^2 - 4x + 3')
plt.plot(x, g, label ='g(x) = -x^3 + 2x^2 + x - 5')
plt.plot(x, h, label ='h(x) = x^4 - 3x^2 + 2')
plt.axis([-2.5,2.5,-100,2000])
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.title("Gráficas de funciones polinómicas")
plt.grid()
plt.show()
```
